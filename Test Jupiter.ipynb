# Ploting test

Here I will plot a function:
$$
y = x^2
$$
<br>
in the range of -10 to 10 with 200 points.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2
x = np.linspace(-10,10,200)
y = f(x)

plt.plot(x, y);
```
